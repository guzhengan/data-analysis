## 4.区间索引

### （a）利用interval_range方法

```python
pd.interval_range(start = 0 ,end = 5) # closed参数可以选择 'left','right','both','neither',默认左开右必 
```

```python
pd.interval_range(start = 0 ,periods = 8 ,freq=5) #periods参数控制区间个数，freq 控制步长
```

### （b）利用cut将数值列转为区间为元素的分类变量，例如统计数学成绩的区间情况：

```python
math_interval = pd.cut(df['Math'],bins = [0,40,60,80,100])
math_interval.head()
```

### (c) 区间索引的选取

```python
df_i = df.join(math_interval,rsuffix='_interval')[['Math','Math_interval']].reset_index().set_index('Math_interval')
df_i.head()
```

```python
df_i.loc[65].head()  # 65 在（60，80] 内， Math列的值在（60，80] 内，都会被选中
```

```python
df_i.loc[[65,90]]
```

#### 如果想要选取某个区间，先要把分类变量转为区间变量，再使用overlap方法

```python
df_i.index
```

```python
# df_i.loc[pd.Interval(70,75)].head()   报错
df_i[df_i.index.astype('interval').overlaps(pd.Interval(70,85))].head()
```

## 二、多级索引
## 1.创建多级索引

### （a）通过from_tuple或from_arrays

1.直接创建元组

```python
tuples = [('A','a'),('A','b'),('B','a'),('B','b')]
mul_index = pd.MultiIndex.from_tuples(tuples,names = ('Upper','Lower'))
mul_index
```

```python
pd.DataFrame({'Score':['perfect','good','fair','bad']},index = mul_index)
```

2.利用zip创建元组

```python
L1 = list('AABB')
L2 = list('aabb')
tuples = list(zip(L1,L2))
mul_index = pd.MultiIndex.from_tuples(tuples,names = ('Upper','Lower'))
pd.DataFrame({'Score':['perfect','good','fair','bad']},index = mul_index)
```

3.通过Array创建

```python
arrays = [['A','a'],['A','b'],['B','a'],['B','b']]
mul_index = pd.MultiIndex.from_tuples(arrays,names = ('Upper','Lower'))
pd.DataFrame({'Score':['perfect','good','fair','bad']},index = mul_index)
```

```python
mul_index # 内部自动转为元组
```

### （b） 通过from_product

```python
L1 = ['A','B']
L2 = ['a','b']
pd.MultiIndex.from_product([L1,L2],names = ['Upper','Lower'])     # 两两相乘
```

## 2.多层索引切片

```python
df_using_mul.head()
```

### （a） 一般切片

```python
df_using_mul.loc['C_1','street_1'] # 当索引未排序时，会性能报错
```

```python
df_using_mul.index.is_lexsorted() # 该函数检查是否排序
```

```python
df_using_mul.sort_index().loc['C_1','street_1']  #先排序再切片
```

```python
df_using_mul.loc[('C_2','street_5'):]  # 未排序时，不能使用多层切片
```

```python
df_using_mul.sort_index().loc[('C_2','street_6'):('C_3','street_4')]  # loc 元素索引，包含右端点
```

```python
df_using_mul.sort_index().loc[('C_2','street_7'):'C_3']  # 一般需要传入元组 ，非元组 单个元素也是可以的，表示选择该层所有元素
```

### （b）第一类特殊情况：由元组构成列表

```python
df_using_mul.sort_index().loc[[('C_2','street_7'),('C_3','street_2')]]  #表示选出某几个元素，精确到最内层索引 
```

### （c）第二类特殊情况：由列表构成元组 

```python
df_using_mul.sort_index().loc[(['C_2','C_3'],['street_4','street_7']),:]
# 选出第一层在‘C_2’和‘C_3’中且第二层在'street_4'和'street_7'的列
```

### 2.reindex和reindex_like

reindex是指重新索引，它的重要特性在于索引对齐，很多时候用于重新排序

```python
df.head()
```

```python
df.reindex(index = [1101,1203,1206,2402])
```

```python
df.reindex(columns=['Height','Gender','Average']).head()
```

#### 可以选择缺失值的填充方法：fill_value和method(bfill / ffill / nearest)，其中method参数必须索引单调

```python
df.reindex([1101,1203,1206,1301],method='bfill') 
#bfill表示用所在索引1206的后一个有效行进行填充，ffill为前一个有效行，nearest指最近一个有效行
```

```python
df.reindex([1101,1205,1206,1301],method='nearest')
```

#### reindex_like的作用为生成一个横纵索引与参数列表一致的DataFrame,数据使用被调用的表

```python
df_temp = pd.DataFrame({'Weight':np.zeros(5),
                        'Height':np.zeros(5),
                       'ID':[1101,1104,1103,1106,1102]}).set_index('ID')
df_temp.reindex_like(df[0:5][['Weight','Height']])  # reindex_like 使用参数df[0:5]作为新DataFrame的索引
```

#### 如果df_temp索引单调，还可以使用填充缺失值的method参数： 

```python
df_temp = pd.DataFrame({'Weight':range(5),
                        'Height':range(5),
                       'ID':[1101,1104,1103,1106,1102]}).set_index('ID').sort_index()
df_temp.reindex_like(df[0:5][['Weight','Height']],method = 'bfill') # 行1105 取得了有效行1106 的值
```

## 五、重复元素处理

### 1.duplicated方法
#### 该方法返回了是否重复的布尔列表

```python
df.duplicated('Class').head(7) # 首次出现 均为False
```

#### 可选参数keep默认为first，即首次出现设为不重复，若为last，则最后一次设为不重复，若为False，则所有重复项为True

```python
df.duplicated('Class',keep = 'last').tail() # 最后一次出现不重复
```

```python
df.duplicated('Class',keep = False).head()
```

### 2.drop_duplicates方法
#### 从名字上看为剔除重复项，分组中可用，例如需要保留每组的第一个值：

```python
df.drop_duplicates('Class')
```

#### 参数与duplicated 函数类似

```python
df.drop_duplicates('Class',keep = 'last')  # 取该组最后一次出现的值
```

#### 在传入多列时等价于将多列共同视为一个多级索引，比较重复项：

```python
df.drop_duplicates(['Class','School'])
```

## 六.抽样函数
#### 这里的抽样函数指的就是sample函数

### （a）n为样本量

```python
df.sample(n=5)
```

### (b) frac 为抽样比

```python
df.sample(frac=0.05) # 结果行数 约为  df.shape[0] * 0.05 即 35*0.05=1.75 约为2
```

### （c）replace为是否放回

```python
df.sample(df.shape[0],replace=True).head()   # 是否可以重复取
```

```python
df.sample(df.shape[0],replace=True).index.is_unique
```

### (d) axis为抽样维度，默认为0，即抽行

```python
df.sample(n=3,axis = 1).head()
```

### (e) weights为样本权重，自动归一化

```python
df.sample(n=3 , weights=np.random.rand(df.shape[0]))
```

```python
# 以某一列为权重，这在抽样理论中很常见
# 抽到的概率与Math数值成正比
df.sample(n=3,weights=df['Math'])
```
