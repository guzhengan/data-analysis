```python
data = df[df['发货地址'].notnull()]
data
```

```python
data[~data['价格'].str.contains(r'^-?(\d+)(\.\d+)?$')]
data = data[data['价格'].str.contains(r'^-?(\d+)(\.\d+)?$')]
#求所有在杭州发货的商品单价均值
data[data['发货地址'].str.contains('杭州')]['价格'].astype('float').mean()
```

```python
data['类别'] = pd.qcut(data['价格'],[0,0.2,0.4,0.6,0.8,1.],labels=['低','较低','中','较高','高'])
data.sort_values('类别', ascending=False)  # 对价格是否需要 .str 数据类型的转换
```
