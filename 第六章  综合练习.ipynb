```python
df1['Winning rate'] = df1['Total number of license issued'] / df1['Total number of applicants']
# df1.head()
df1[df1['Winning rate']<0.05]['Date'].values[0]
```

(3) 将第一列时间列拆分成两个列，一列为年份（格式为 20××），另一列为

```python
df2 = pd.read_csv('/Users\gan\Desktop/2007年-2019年俄罗斯货运航班运载量.csv')
df2.head()
```

#### (1) 求每年货运航班总运量。

```python
df2.groupby('Year')['Whole year'].sum()
```

####  (2) 每年记录的机场都是相同的吗？

```python
df2.groupby('Year')['Airport name'].count()
```

#### (3) 按年计算 2010 年-2015 年全年货运量记录为 0 的机场航班比例。

```python
df2_3 = df2.set_index(['Year']).sort_index()
df2_3.loc[2010:2015]['Whole year'].groupby('Year').agg(lambda x:print( 
    '年份：', x.index[0], '\t 比例：%.2f' % (len(x[x==0]) / len(x) * 100), '%' ) )
```

#### (4) 若某机场至少存在 5 年或以上满足所有月运量记录都为 0，则将其所有年份的记录信息从表中删除，并返回处理后的表格

```python
df2_4_group = df2.set_index('Airport name').groupby('Airport name')
temp0 = df2_4_group['Whole year'].apply( lambda x: len(x[x==0])>5) #数据列格式
temp0 = pd.DataFrame(temp0) #不转换 下面要报错
display(temp0.info()) #为True表示 要删掉的那些。他们在temp0中有序、分组排列
df2.set_index('Airport name').drop(temp0[temp0['Whole year']==True].index)
```
