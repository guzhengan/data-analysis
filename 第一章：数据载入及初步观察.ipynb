#### 数据清洗时，会将带空值的行删除，此时DataFrame或Series类型的数据不再是连续的索引，可以使用reset_index()重置索引。

```python
df_e = pd.DataFrame(np.arange(20).reshape(5,4),index=[1,3,4,6,8])
df_e
```

```python
df_e.reset_index()
```

```python
df_e.reset_index(drop = True)   # 在获得新的index，原来的index变成数据列，保留下来。
                               # 不想保留原来的index，使用参数 drop=True，默认 False。
```

#####  1.6.1 任务一：利用Pandas对示例数据进行排序，要求升序

```python
df_e3 = pd.DataFrame(np.arange(8).reshape((2,4)),index = [2,1],columns = ['d','a','b','c'])
df_e3
```

##### 【问题】：大多数时候我们都是想根据列的值来排序,所以将你构建的DataFrame中的数据根据某一列，升序排列

```python
df_e3.sort_values(by = 'c')  # 默认升序
```

```python
df_e3.sort_values(by = 'c',ascending = False) 
```

```python
# 让行索引升序排序
df_e3.sort_index()
```

```python
# 让列索引升序排序
df_e3.sort_index(axis=1)
```

```python
# 让列索引降序排序
df_e3.sort_index(axis=1,ascending=False)
```

```python
# 让任选两列数据同时升序排序
df_e3.sort_values(by=['a', 'c'])
```
