### 2 .筛选的逻辑--表格数据中，最重要的一个功能就是要具有可筛选的能力，选出我所需要的信息，丢弃无用的信息。

#### 2.2.1 筛选出 年龄Age小于 10 的 乘客

```python
df_train[df_train['Age']<10].head()
```

#### 2.2.2 以"Age"为条件，将年龄在10岁以上和50岁以下的乘客信息显示出来，并将这个数据命名为midage

```python
condition = (df_train.Age>10) & (df_train.Age<50)
midage = df_train[condition]
midage    # 条件筛选  & | not 的使用 
```

#### 2.2.3 将midage的数据中第100行的"Pclass"和"Sex"的数据显示出来

```python
midage = midage.reset_index()
midage   # reset_index 的作用： 根据上面的筛选条件筛出的数据，存在部分行空缺的情况，如：行索引 7/10/16 缺失
          # 如果不reset_index() 将对后续 数据的查找（索引操作）存在影响
         # reset_index的参数 drop = True 即把原来缺失的索引（这里的index删除）,默认drop = False 不删除
```

```python
midage = midage.reset_index(drop = True)
midage
```

```python
midage.loc[[99],['Pclass','Sex']]     # 这里需要注意 [ ] 的数量，当某行或某列是多个值时，传入行和列的列表！！
```

```python
midage.loc[99]['Sex']
```

#### 2.2.4 使用loc方法将midage的数据中第100，105，108行的"Pclass"，"Name"和"Sex"的数据显示出来

```python
midage.loc[[99,105,108],['Pclass','Name','Sex']]
```

#### 2.2.5 使用iloc方法将midage的数据中第100，105，108行的"Pclass"，"Name"和"Sex"的数据显示出来

```python
midage.iloc[[99,105,108],[2,3,4]]     #这里的  iloc 方法，注意 这个 i 指的是index 索引   （iloc位置索引）
                                # 后面的切片 iloc 方法 不包含尾端，而loc方法则包含尾端，
                               # 因为既然我们知道了loc 标签索引，那么我们就需要知道它的值
```

#### 3.1.1 利用Pandas对示例数据进行排序，要求升序

```python
# 首先创建一个较简单的DataFrame数据例子
df_example = pd.DataFrame(np.arange(8).reshape((2,4)),
                         index = ['2','1'],
                         columns = ['d','a','b','c'])   # 创建一个2*4的二维数组，范围arange（8），
                                                        # 注意reshape((2,4)) 两个括号
                                                        # 行索引 ['2','1'] ，列索引['d','a','b','c']
df_example
```

```python
# 大多数时候我们都是想根据列的值来排序,所以，将你构建的DataFrame中的数据根据某一列，升序排列
df_example.sort_values(by='c', ascending=True)   # 排序函数  sort_values ( by =,ascending = True)
               # by 参数，指定某列，ascending 升序or降序，默认=True 是升序
              # 实际传参还有多种方式，如：by = 'c' 或者 by = ['a','c'] 
```

```python
# by =    参数多个值 的例子
df_example.sort_values(by=['a','c'], ascending=False)  # 优先 传入by参数的第一个值 
```

```python
# 让行索引升序排序
df_example.sort_index()
```

```python
# 让列索引升序排序
df_example.sort_index(axis=1)
```

```python
# 让列索引降序排序
df_example.sort_index(axis=1, ascending=False)
```

```python
# 让任选两列数据同时降序排序
df_example.sort_values(by=['a', 'c'], ascending=False)
```

#### 3.3 利用Pandas进行算术计算，计算两个DataFrame数据相加结果

```python
#创建数据：
frame1_a = pd.DataFrame(np.arange(9.).reshape(3, 3),
                     columns=['a', 'b', 'c'],
                     index=['one', 'two', 'three'])
frame1_b = pd.DataFrame(np.arange(12.).reshape(4, 3),
                     columns=['a', 'e', 'c'],
                     index=['first', 'one', 'two', 'second'])
```

```python
frame1_a
```

```python
frame1_b
```

```python
# 直接将两个 DataFrame 进行相加
frame1_a + frame1_b      #   由此发现 只有行索引和列索引完全相同的情况下，数值才会进行计算，其他则为NaN
```
